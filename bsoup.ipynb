```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
data = requests.get(url)
```

```python
bsoup = BeautifulSoup(data.text)
table = bsoup.select('table.stats_table')[0]
```

```python
links = table.find_all('a')
```

```python
squads = [i.get('href') for i in links]
```

```python
squads = [l for l in squads if '/en/squads/' in l]
```

```python
squads_urls = [f"https://fbref.com{j}" for j in squads]
```

```python
squads_urls
```

```python
team_url = squads_urls[0]
```

```python
data = requests.get(team_url)
```

```python
matches = pd.read_html(data.text, match="All Competition")[0]
```

```python
matches
```

```python
soup = BeautifulSoup(data.text)
```

```python
comps = soup.find_all('a')
```

```python
links = [i.get('href') for i in comps]
```

```python
links
```

```python
comps_all = [i for i in links if i and 'all_comps/shooting/' in i]
```

```python
comps_all
```

```python
data2 = requests.get(f"https://fbref.com{comps_all[0]}")
```

```python
shooting = pd.read_html(data2.text, match='Shooting')[0]
```

```python
shooting.columns = shooting.columns.droplevel()
```

```python
shooting.head()
```

```python
team_data = matches.merge(shooting[['Date', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']], on="Date")
```
